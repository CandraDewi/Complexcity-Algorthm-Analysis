**Studi Case**
- Menganalisis pengaruh penggunaan struktur data yang berbeda terhadap efisiensi (dalam hal kompleksitas waktu) dari algoritma pencarian yang diterapkan pada sistem perpustakaan digital. Fokus utama adalah mengukur dan membandingkan performa beberapa algoritma pencarian ketika diimplementasikan pada berbagai struktur data, dengan studi kasus pencarian buku dalam database perpustakaan digital yang berisi informasi seperti judul buku, penulis, ISBN, tahun terbit, dan kategori.

```python
# Import Library
import time # untuk mengukur waktu eksekusi algoritma
import sys # untuk mengatur batas rekursi
import matplotlib.pyplot as plt # untuk membuat grafik performa
from tabulate import tabulate # untuk mencetak tabel waktu eksekusi

# Mengatur batas rekursi
sys.setrecursionlimit(1500)

# Fungsi Iterative Search
def iterative_search(array, target):
    for i in range(len(array)):
        if array[i] == target:
            return i
    return -1

# Fungsi Rekursif Search
def recursive_search(array, target, index=0):
    if index >= len(array):
        return -1
    if array[index] == target:
        return index
    return recursive_search(array, target, index + 1)

# Fungsi untuk mencetak tabel waktu eksekusi
def print_execution_table(indices, iterative_times, recursive_times):
    table_data = []
    for i in range(len(indices)):
        table_data.append([indices[i], f"{iterative_times[i]:.6f}s", f"{recursive_times[i]:.6f}s"])
    print("\nExecution Time Table:")
    print(tabulate(table_data, headers=["Dataset Size (n)", "Iterative Time", "Recursive Time"], tablefmt="grid"))

# Program  utama
def main():
    # User input for dataset size
    n = int(input("Enter the number of books (n): "))

    # Generate dataset
    books = [f"Book {i}" for i in range(1, n + 1)]

    iterative_times = []
    recursive_times = []
    indices = list(range(1, n + 1, max(1, n // 10)))  # Evaluate for 10 points or fewer

    # Mengukur dan menyimpan waktu eksekusi
    for i in indices:
        target_book = f"Book {i}"

        # Perkiraan waktu menggunakan algoritma iteratif
        start_time = time.time()
        iterative_search(books, target_book)
        iterative_times.append(time.time() - start_time)

        # Perkiraan waktu menggunakan algoritma rekursif
        start_time = time.time()
        recursive_search(books, target_book)
        recursive_times.append(time.time() - start_time)

        # Mencetak waktu eksekusi untuk setiap iterasi
        print(f"For n = {i}: Iterative = {iterative_times[-1]:.6f}s, Recursive = {recursive_times[-1]:.6f}s")

    # Cetak tabel waktu eksekusi
    print_execution_table(indices, iterative_times, recursive_times)

    # Grafik hasil perbandingannya
    plt.figure(figsize=(10, 6))
    plt.plot(indices, iterative_times, label='Iterative', marker='o', color='blue')
    plt.plot(indices, recursive_times, label='Recursive', marker='o', color='green')
    plt.title('Performance Comparison of Iterative vs Recursive Search')
    plt.xlabel('Dataset Size (n)')
    plt.ylabel('Execution Time (seconds)')
    plt.legend()
    plt.grid(True)

    # Menyimpan dan menampilkan grafik perbandingan
    plt.savefig('performance_comparison_line_chart.png')
    plt.show()

if __name__ == "__main__":
    main()
```

**Kesimpulan:**
- Algoritma Iteratif lebih efisien untuk pencarian dalam dataset berukuran besar.
- Algoritma Rekursif memiliki performa serupa untuk dataset kecil tetapi menjadi kurang efisien seiring pertumbuhan dataset karena batasan rekursi dan overhead stack.
